```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 1: Create data manually (FIXED)
data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [100, 120, 130, 150, 170, 180,
              200, 210, 230, 250, 270, 300]
})

# Step 2: Show dataset
print("Dataset Preview:")
print(data.head())

# Step 3: Data preprocessing
data['Month_num'] = range(1, len(data) + 1)

X = data[['Month_num']]
y = data['Sales']

# Step 4: Train model
model = LinearRegression()
model.fit(X, y)

# Step 5: Model evaluation
predictions_train = model.predict(X)
mse = mean_squared_error(y, predictions_train)
print("\nModel Evaluation:")
print("Mean Squared Error:", mse)

# Step 6: Future prediction
future_months = pd.DataFrame({'Month_num': [13, 14, 15]})
future_predictions = model.predict(future_months)

print("\nPredicted Sales for next 3 months:")
for i, val in enumerate(future_predictions, start=13):
    print(f"Month {i}: {val:.2f}")

# Step 7: Visualization
plt.figure()
plt.scatter(X, y, label="Actual Sales")
plt.plot(X, predictions_train, label="Regression Line")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.title("Sales Forecast using Linear Regression")
plt.legend()
plt.show()

# Step 8: Business Insight
print("\nBusiness Insight:")
print("Sales show an increasing trend over time. The model predicts continued growth.")
```
